```python
import pandas as pd
import sys, os
import os.path as osp
from datetime import datetime
```

```python
df = pd.read_csv('../data/nyc_311_limit.csv', header=None)
df
```

```python
#remove rows with no zipcode or end date
df = df.dropna(axis=0, subset = [2, 8])
df
```

```python
#remove rows with wrong zip
df = df[pd.to_numeric(df[8], errors='coerce').notnull()]
df
```

```python
df[1] = df[1].apply(lambda x: if !isinstance(x, datetime.datetime): datetime. strptime(x, '%m/%d/%Y %I:%M:%S %p'))
df[2] = df[2].apply(lambda x: if !isinstance(x, datetime.datetime): datetime. strptime(x, '%m/%d/%Y %I:%M:%S %p'))
df
```

```python
df_2020 = df[df[1].dt.year==2020]
df_2020
```

```python
#Remove data with close date < open date
df_2020 = df_2020[df_2020[2] >= df_2020[1]]
df_2020
```

```python
a,_= df_2020.shape
b,_ = df.shape
print(f'2020: {a}; no: {b}')
```
