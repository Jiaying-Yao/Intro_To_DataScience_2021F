```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("../data/nyc_311_limit_2020.csv", header = None)
df
```

```python
df.info()
```

```python
df[1] = df[1].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))
df[2] = df[2].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))

df['Response_time'] = (df[2] - df[1]).dt.total_seconds()
df['Response_time'] = df['Response_time'].apply(lambda x: divmod(x, 3600)[0])
df
```

```python
df['Month'] = df[2].dt.month
df
```

```python
df1 = df
#df1 = df.groupby([[8, 'Month']], as_index=False).mean()
#df1 = df1[[x, 'Response_time']]
df1 = df1[[8,'Month','Response_time']]
```

```python
df1 = df1.rename(columns = {8: 'zipcode'})
df1
```

```python
df1.to_csv("../data/extract_response.csv", index = False)
```

```python
df2 = df1.pivot_table(index='zipcode', columns = 'Month')
df2.rename_axis(None)
df2
```

```python
df2['Month'] = df2.index
df2
```
